```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.fixmystreet.com/reports')
soup = BeautifulSoup(page.content,'html.parser')
soup = soup.find(class_ = 'js-autocomplete')
```

```python
options = soup.findAll("option")
councils = [y.text for y in options]
councils = councils[1:]
councils
```

```python
new_councils = []
final_councils = []

for council in councils:
    council = str(council)
    words = council.split(' ')
    
    try:
        index = words.index('Council')
        words = words[:index]
        new_councils.append(words)
    except:
        pass
    
for council in new_councils:
    
    if 'City' in council or 'Borough' in council or 'District' in council:
        council = council[:-1]
    else:
        pass
    final_councils.append(council)

```

```python
url_council = []
for councils in final_councils:
    z = ''

    for i in range(len(councils)):
        z = z + str(councils[i]) + '+' 
    url_council.append(z[:-1])
url_council
```

```python
urls = []
for council in url_council:
    url = 'https://www.fixmystreet.com/reports/' + str(council)
    urls.append(url)

```

```python
data = []

for url in urls:
    for i in range(20):
        try:
            page = requests.get(url + '?p=' + str(i+1))
            soup = BeautifulSoup(page.content,'html.parser')
            scrap = soup.find(class_ = 'item-list item-list--reports')
            scrap = scrap.find_all('a')
            report_links = []
            for link in scrap:
                report_links.append(link.attrs['href'])
            for reports in report_links:
                reportpage = requests.get(reports)
                reportsoup = BeautifulSoup(reportpage.content,'html.parser')
                reportsoup = reportsoup.find('div','moderate-display')
                data.append(reportsoup.text)
        except:
            pass
```
